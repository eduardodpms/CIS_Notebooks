## **4. Análise Estatística e Elaboração do Modelo**

### **4.1. Importação e Tratamento do Dataset**

A primeira etapa do *pipeline* desenvolvido, conforme explicado no tópico [*3.3. Execução de Pipeline Próprio*](3-3-execução-de-pipeline-próprio), consistirá na importação e tratamento do *Dataset* fornecido. Inicia-se o processo com a importação das bibliotecas necessárias:

```python
import pandas as pd # Manipulação de arquivos
import matplotlib.pyplot as plt # Geração de figuras
import seaborn as sns # Visualização estatística
import numpy as np # Manipulação de arrays
```

Depois, segue-se para a importação do *Dataset*, que será realizada com o auxílio da biblioteca [*Pandas*](https://pandas.pydata.org). Além disso, por questões de visualização, é conveniente utilizar o método *to_string* para uma rápida observação e entendimento do arranjo da relação.

```python
data = pd.read_csv(filepath_or_buffer='diabetes.csv') # Tranfere-se o conjunto de dados para um Dataframe, com o auxílio da biblioteca Pandas

print(f'{data.to_string(max_rows=10)}') # Apresenta as n primeiras e n últimas linhas da relação
```

Após essa observação do modelo do *Dataset*, é possível destacar alguns pontos importantes:
- A relação está dividida em 768 (0-767) tuplas e 9 atributos;
- 3 atributos apresentam apenas valores inteiros ("*Pregnancies*", "*Age*" e "*Outcome*"), enquanto os outros são valores de ponto flutuante;
- A análise central deverá ser focada no atributo "*Outcome*", que representa o diagnóstico (1) ou não (0) de *diabetes mellitus*;
- É perceptível a presença de valores nulos (= 0) em atributos como "*Pregnancies*", "*SkinThickness*" e "*Insulin*".

Este último tópico aponta para a necessidade de um tratamento do *Dataset*. Enquanto as variáveis "*Pregnancies*" e "*Outcome*" possuem uma lógica por trás que as permite assumir um valor nulo, todos os outros índices, pela própria natureza de medição clínica que eles expressam e pela maneira que são calculados, não podem, nas condições de operação do corpo humano (saudável ou não), manifestar um valor absoluto nulo. Vale ressaltar, ainda, que isso não os impede de assumir valores extremamente pequenos.

Isso significa que, para as variáveis além de "*Pregnancies*" e "*Outcome*", os dados nulos estão inconsistentes. São um tipo dos chamados "*Outliers*", pois distoam do resto do conjunto de dados, e não é incorreto concluir, em uma análise estatística como essa, que estão incorretos. Seguindo uma lógica parecida, é uma boa prática analisar *Datasets* em busca de valores faltantes, que muitas vezes aparecem nas relações como "*Null*". Para isso, é comum a utilização do método "*info*". Aqui, será extendida a função desse método para que também mostre a contagem de valores nulos, com a criação da seguinte função:

```python
# Função que opera similarmente ao método "info", do Pandas Dataframe
def print_info(data):
  print(f"RangeIndex: {len(data)} entries, 0 to {len(data)-1}")
  print(f"Columns: total {data.shape[1]} columns\n")
  print(f"{' #':>3} {'Column':<15} {'Null Count':>15} {'Zero Count':>15} {'Dtype':>10}")
  print("-" * 62)

  for i, col in enumerate(data.columns):
      dtype = data[col].dtype
      non_null = data[col].isnull().sum()

      if np.issubdtype(dtype, np.number):
          non_zero = (data[col] == 0).sum()
      else:
          non_zero = 'N/A'

      print(f"{i:>3} {col:<15} {non_null:>15} {str(non_zero):>15} {str(dtype):>10}")
```

Criada a função que auxiliará a análise antes do tratamento em si, segue-se para a sua aplicação:

```python
data.rename(columns={'DiabetesPedigreeFunction': 'DPF'}, inplace=True) # Por questões de visualização, abrevia-se o índice "DiabetesPedigreeFunction" para "DPF"

print_info(data) # Visualização da contagem de dados nulos e de dados iguais a zero
```

Com o uso da função personalizada para visualização da contagem de valores *Null* e iguais a zero, é perceptível que as inconsistências supostas no parágrafo anterior ao anterior se extendem para os atributos "*Glucose*", "*BloodPressure*" e "*BMI*". Ainda, percebe-se que a métrica "*DPF*", possui um único valor *Null*.

Para uma melhor visualização da distribuição dos dados para cada atributo, é apropriada a utilização de histogramas:

```python
plot = data.hist(figsize = (20,12), layout=[4,5]) # Cria um histograma para cada atributo do conjunto de dados
```

Aqui, prossegue-se para uma análise visual dos histogramas apresentados. As variáveis em interesse para o tratamento são "*Glucose*", "*BloodPressure*", "*SkinThickness*", "*Insulin*" e "*BMI*". É claramente perceptível, com excessão de "*Insulin*", que seus gráficos se aproximam de uma curva de *Gauss*[⁽⁵⁾](#6-referências), isto é, uma curva normal, quando desconsideramos as anormalidades causadas pelos valores iguais a zero. Isso significa, resumidamente, que o histograma tende a ser simétrico na média e que a maioria dos valores estão concentrados ao redor dela.

Por causa desse fenômeno, uma maneira eficiente e simples de se tratar os dados incongruentes sem alteração do "centro de massa" do gráfico (que se encontra na média), é a substituição pela própria média. Esse é um dos diversos métodos[⁽⁶⁾](#6-referências) de "Imputação Estatística", que consiste na maneira como se trata os dados nulos e *Outliers* em *Datasets*.

Com isso, aplica-se a Imputação Estatística nos atributos de interesse do *Dataset* do *NIDDK*. Ressaltam-se, ainda, dois pontos importantes:
- O atributo "*Insulin*", ou nível de Insulina no sangue, pode levantar suspeitas quanto à natureza de seu gráfico, pois não é uma boa prática assumir que ele seria normal se fossem desconsiderados os dados inconsistentes. Para os fins deste trabalho, será tomado que seu histograma também é aproximadamente normal, uma vez que a curva de *Glucose*, ou taxa de glicose, é evidentemente normal, e há uma relação direta[⁽⁴⁾](#6-referências) entre esses atributos;
- Conforme visualizado na contagem de dados *Null* e iguais a zero, na célula de texto anterior, o atributo *DPF* possui um único valor *Null*, provavelmente advindo de algum erro desconhecido. Substituí-lo pela média não trará grandes preocupações para a análise aqui desenvolvida, e assim será feito a fim de proporcionar um estudo simples e conciso.

```python
cols_zero = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI'] # Separando os atributos aproximadamente normais que possuem valores nulos

data_new = data.copy(deep = True) # Salvando o Dataframe que sofrerá tratamento em "data_new"
```

A primeira etapa da Imputação Estatística será a substituição dos valores iguais a zero por *Null* em todos os atributos de interesse, com o uso do método *replace*. Depois, aplica-se a substituição pela média com o método *fillna*.

```python
data_new[cols_zero] = data_new[cols_zero].replace(0,np.nan) # Substitui os valores "zero" dos atributos por NaN

for col in cols_zero + ["DPF"]: # Adiciona o atributo "DPF", para consertar seu valor nulo
  data_new[col] = data_new[col].fillna(data_new[col].mean()) # Substitui os valores "NaN" pela média do atributo
```

Com o conjunto de dados devidamente tratado, é conveniente verificar como se encontra a contagem de valores *Null* e iguais a zero:

```python
print_info(data_new) # Visualização da contagem de dados nulos e de dados iguais a zero
```

Depois, para concluir a etapa de tratamento de dados, realiza-se uma validação dos histogramas para os atributos alterados:

```python
plot = data_new.hist(figsize = (20,12), layout=[4,5]) # Cria um histograma para cada atributo do conjunto de dados
```

Aqui, é perceptível como a Imputação Estatística pela média manteve certa "normalidade" nos histogramas que sofreram alteração.

### **4.2. Análise do *Dataset* (tratado)**

Antes de iniciar o desenvolvimento do modelo, é relevante proceder com uma rápida análise das relações existentes entre cada um dos atributos. Para isso, utiliza-se de dois métodos principais, fornecidos pela biblioteca [*seaborn*](https://seaborn.pydata.org): o *pairplot* e o *heatmap*.

Primeiramente, observa-se a *plotagem* de gráficos utilizando o *pairplot*:

```python
sns.pairplot(data=data_new)
```

O *pairplot* é uma função que fornece uma riqueza de análises para um abordagem que traga uma quantidade alta de variáveis, uma vez que permite a visualização de histogramas (presentes na diagonal da imagem) e de gráficos que mostram a relação entre cada par de atributos.

Normalmente, o *pairplot* é muito utilizado quando a variável de interesse (nesse caso, o *Outcome*) é quantitativa, uma vez que o estudo da distribuição dos pontos em cada gráfico permite a dedução de conclusões sobre a relação entre os atributos. Um exemplo de relação é a linear, na qual o arranjo dos pontos se dá em torno de uma reta, cujos coeficientes podem ser obtidos pela técnica de Regressão Linear[⁽⁷⁾](#6-referências).

Todavia, a variável *Outcome*, em torno da qual está organizado este estudo, é uma variável qualitativa (0 = sem diabetes, 1 = com diabetes). Nesse sentido, percebe-se a maneira como a investigação das distribuições no *pairplot* que envolvam esse atributo não leva a conclusões significativas. Por essa razão, a Regressão Linear[⁽⁷⁾](#6-referências) não será a escolha preferível para a abordagem desenvolvida.

Com isso, uma alternativa altamente fundamentada é a chamada "Correlação de *Pearson*"[⁽⁸⁾](#6-referências). Trata-se de uma métrica que mensura a direção e o grau da relação linear entre duas variáveis em uma escala de *-1* até *1*, em que "**1**" seria o caso em que uma variável possui relação direta com outra, "**-1**" seria o caso de uma relação inversa, e "**0**" seria um caso sem qualquer relação. Aqui, ela será utilizada em conjunto com um *heatmap* do [*seaborn*](https://seaborn.pydata.org) para uma visualização clara e simples da relação entre todos os pares de variáveis.

```python
plt.figure(figsize=(7,6))
sns.heatmap(data_new.corr(), annot=True, cmap='rocket_r')
```

Com a utilização desse coeficiente em conjunto com o *heatmap*, evidencia-se a relação entre todos os atributos. Destacam-se as seguintes:

- Idade e Gestações (0.54);
- IMC e Expessura da Pele (0.53);
- Glicose e *Outcome* (0.49);
- Glicose e Insulina (0.41);

Dessa forma, conclui-se, no escopo desta abordagem, que a variável que mais possui relação com o diagnóstico de *diabetes mellitus* é o nível de glicose no sangue. Trata-se de um resultado esperado, conforme apresentado pelo Ministério da Saúde[⁽⁴⁾](#6-referências).

### **4.3. Elaboração do Modelo**

Após a etapa de importação, tratamento e análise do *dataset*, prossegue-se para o desenvolvimento de um modelo preditivo.

O K-Nearest Neighbors (KNN)[⁽⁹⁾](#6-referências) é um modelo com aprendizado supervisionado que é tipicamente utilizado como um algoritmo de classificação para *datasets* considerados "pequenos". De forma resumida, ele funciona trabalhando com uma média dos *K* pontos vizinhos (ou K-nearest neighbors) em um determinado ponto de consulta para fazer uma previsão. O valor de *K* aqui é um parâmetro fundamental e impactará diretamente no desempenho do modelo, conforme será visto no tópico [*4.7. Otimização do Modelo*](#4-7-otimização-do-modelo).

Nesse sentido, o KNN é um algoritmo de "aprendizado preguiçoso", isto é, ao contrário de modelos de aprendizado mais complexos, que passam pela fase de treinamento característica, com emprego de camadas de percéptrons e funções de ativação, ele apenas armazena o conjunto de dados de treinamento. Por essa razão, trabalha melhor com *datasets* "pequenos", como o **Pima Indians Diabetes Database** [⁽¹⁾](#6-referências). Além disso, os atributos de treino desse *dataset* são quantitativos, o que significa que podem ser normalizados, algo crucial para a escalonação das distâncias em um gráfico. Assim, o KNN pode entregar uma performance vantajosa e com pouco custo computacional.

Inicia-se o processo de elaboração do modelo com a definição das principais variáveis e a importação das bibliotecas do [*scikit-learn*](https://scikit-learn.org).

Observação: o valor "ótimo" da variável *K* (n_neighbors) foi definido após a etapa [*4.7. Otimização do Modelo*](#4-7-otimização-do-modelo).

```python
n_neighbors = 14 # Valor de K que será utilizado no KNN
test_size = 0.2 # Proporção de dados de teste em relação ao tamanho do dataset
random_state = 42 # Como o modelo aleatorizará o dataset
```

```python
# Importação das bibliotecas de classificação do Scikit Learn
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, accuracy_score
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

Aqui, separa-se o *dataset* nas variáveis *X* (*input*) e *y* (*output*):

```python
# Separando features e target
x = data_new.drop('Outcome', axis=1)
y = data_new['Outcome']
```

Depois, avança-se para o processo de **normalização**, que é fundamental no treinamento do modelo. A normalização faz com que os valores de diferentes atributos variem dentro de um mesmo intervalo (nesse caso, *μ* = 0 e *σ* = 1), sem trazer prejuízos de escala para os cálculos executados pelo modelo. Utiliza-se a classe *StandartScaler*, do [*scikit-learn*](https://scikit-learn.org), para essa tarefa.

```python
# Normalizando os dados
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

### **4.4. Treinamento do Modelo**

Com o conjunto de features (dados de treino) normalizados, entra em ação a etapa de treinamento do modelo.

Primeiramente, dividem-se os dados em conjuntos de treino e de teste, na proporção definida anteriormente:

```python
# Separando dados de treino e de teste
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size=test_size, random_state=random_state)

print(f"Train/Test Size: {len(x_train)}/{len(x_test)}")
```

Agora, o treinamento é efetivado com a classe *KNeighborsClassifier* e o método *fit*, além do valor de *K* definido anteriormente.

```python
# Treinando o modelo KNN
knn = KNeighborsClassifier(n_neighbors=n_neighbors)
knn.fit(x_train, y_train)
```

### **4.5. Avaliação do Modelo (resultados)**

Depois da etapa de treinamento, testa-se a predição do modelo com o método *predict*, criando o vetor de predições "*y_pred*":

```python
# Testando o modelo
y_pred = knn.predict(x_test)
```

Por fim, compara-se o vetor de predições com o vetor de *labels* em (*y_test*) para obter as métricas de acurácia do modelo:

```python
# Métricas do teste
print(f"Acurácia: {accuracy_score(y_test, y_pred)}\n")
print(classification_report(y_test, y_pred))
```

Nesse caso, obteve-se uma acurácia de aproximadamente **76%** para um *K* = 14. Isso significa que por volta de 76% das respostas predizidas pelo modelo com base nos 8 atributos de entrada corresponderão à realidade.

Uma outra forma eficiente de visualizar os resultados é através de uma matriz de confusão, disposta a seguir com o auxílio da [*Matplotlib*](https://matplotlib.org):

```python
print(f"Test Size: {len(x_test)}")

# Matriz de confusão
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=[0, 1])
disp.plot(cmap=plt.cm.Blues)
plt.title(f"KNN's Confusion Matrix (K={n_neighbors})")
plt.grid(False)
plt.show()
```

A matriz de confusão traz uma abordagem vantajosa ao dividir as predições em quadro quadrantes:

- ✅ Verdadeiro Positivo (Quadrante Inferior Direito) - 32;
- ✅ Verdadeiro Negativo (Quadrante Superior Esquerdo) - 85;
- ❎ Falso Positivo - Erro Tipo 1 (Quadrante Superior Direito) - 14;
- ❎ Falso Negativo - Erro Tipo 2 (Quadrante Inferior Esquerdo) - 23.

<br>

Dessa maneira, conclui-se a etapa de avaliação do modelo. Por fim, é importante destacar algumas observações do modelo:

- A acurácia obtida (~ 76%), é uma acurácia considerada "boa", dadas as condições fornecidas;

- O modelo provavelmente não poderá alcançar acurácias superiores a essa, devido à natureza de seu funcionamento em função do valor de *K*, conforme será visto no tópico [*4.7. Otimização do Modelo*](#4-7-otimização-do-modelo);

- O modelo obteve maior sucesso, conforme visto na matriz de confusão, na identificação de pessoas **sem diabetes**. Um motivo possível para isso é a quantidade maior de dados de treino referentes à indivíduos sem diabetes (500);

### **4.7. Otimização do Modelo**

Este último tópico é dedicado à otimização do modelo que, no caso do *K-Nearest Neighbors*, se dá pela alteração do número de "vizinhos" avaliados no cálculo da predição, ou *K*. A melhor forma de avaliar o efeito de diferentes valores de *K* na acurácia, levando em conta o curto período de tempo que o modelo leva para ser treinado, é repetindo o processo de treino para cada *K* em um intervalo comumente aceito.

Vale perceber que valores muito altos de *K* podem levar a uma menor sensibilidade do modelo a padrões, enquanto valores muito baixos podem levar a *overfitting*, conforme apresentado pela *IBM*[⁽⁹⁾](#6-referências). Dessa forma, é sempre preferível um valor de *K* "mediano" que traga uma baixa taxa de erro. Assim, define-se um intervalo com 30 valores de *K* (1-30) para uma análise extensa.

```python
error_rates = [] # Armazena os valores de erro (1-acurácia) para cada K
k_values = range(1, 31) # Define o range de valores de K
```

Aqui, o loop executará o processo de treinamento e gravará o erro para *KNN*'s com diferentes *K*:

```python
# Examinando o KNN para cada K discreto
for k in k_values:
    knns = KNeighborsClassifier(n_neighbors=k)
    knns.fit(x_train, y_train)
    pred_k = knns.predict(x_test)
    acc = accuracy_score(y_test, pred_k)
    error = 1 - acc
    error_rates.append(error)
```

Por fim, com o auxílio da biblioteca [*Matplotlib*](https://matplotlib.org), é plotado um gráfico que evidencia o valor de erro para cada *K* inteiro, no intervalo de 1 a 30:

```python
# Plotando a taxa de erro para cada K discreto
plt.figure(figsize=(7, 4))
plt.plot(k_values, error_rates, color='blue', linestyle='dashed', marker='o', markerfacecolor='red')
plt.title('Curva de Erro para K')
plt.xlabel('K')
plt.ylabel('Taxa de erro')
plt.grid(True)
plt.show()
```

Com a análise da curva de erro apresentada, é perceptível que há alguns pontos de mínimo - nos quais a abordagem estará centrada - tais como *K* = 5, *K* = 14, *K* = 19, *K* = 22 e *K* = 30. Entretanto, pelo forma como as *KNN*'s se comportam[⁽⁹⁾](#6-referências) com *Datasets* pequenos, é esperado que não só esses pontos de mínimo assumam o mesmo valor, como também que o modelo dificilmente assuma um erro menor que esse.

Com isso, o estudo se limitará, pela razão de intervalos extensos favorecerem comportamentos indesejados e de os valores de erro mostrarem uma tendência a compreenderem um valor mínimo, ao intervalo sugerido no início do tópico. Assim, um valor mínimo que se aproxima do centro desse intervalo ocorre quando *K* = 14, expressando o erro um valor aproximado de 24% (isto é, uma acurácia de 76%).

Por esse motivo, o valor de *K* trabalhado no tópico [*4.3. Elaboração do Modelo*](#4-3-elaboração-do-modelo) é **14**.

### ***5.5. Questão 05***

*Existe uma relação entre o IMC dos pacientes e o diagnóstico de diabetes?
Compare os valores médios de IMC entre os grupos com e sem diabetes, e
analise a diferença estatisticamente.*

Primeiramente, é importante separar os valores de média para esses dois grupos antes da análise:

```python
# Separando os conjuntos
zero_bmi = data_new[data_new["Outcome"] == 0]["BMI"]
one_bmi = data_new[data_new["Outcome"] == 1]["BMI"]

# Separando as médias
zero_mean = zero_bmi.mean()
one_mean = one_bmi.mean()

# Visualizando as médias
print(f'Média para Outcome = 0: {zero_mean}')
print(f'Média para Outcome = 1: {one_mean}')
```

Uma das maneiras estatísticas de se analisar a diferença entre médias consiste no *Teste-T de Amostras Independentes*[⁽¹⁰⁾](#6-referências), também conhecido como *Teste t de Student* ou *t-test*. De forma resumida, ele compara a média entre dois grupos para avaliar estatisticamente o quão significativa ela é para o estudo. Seu funcionamento se baseia em dois resultados:

- O valor *p* avalia a significatividade da diferença, para *p* < 0.05 (sigificativa) e *p* ≥ 0.05 (não-significativa);
- O valor *t*, que representa o tamanho da diferença entre as médias em relação à variabilidade dos dados.

Utiliza-se a biblioteca [*ScyPy*](https://scipy.org) para o cálculo dessas métricas:

```python
from scipy.stats import ttest_ind # Biblioteca para métricas do t-test

t_value, p_value = ttest_ind(zero_bmi, one_bmi, equal_var=False)  # Aplicando o t-test

print(f'Valor t: {t_value}\nValor p: {p_value}') # Visualizando os resultados
```

Assim, com um módulo de *t* relativamente alto, indicando uma diferença grande no que tange à variabilidade do conjunto de dados, e um valor de *p* extremamente baixo, indicando que essa diferença é muito relevante para a abordagem, é possível concluir que existe sim uma relação entre o *IMC* e o diagnóstico de diabetes no contexto desse *dataset*. Ressalta-se, ainda, que o valor negativo de *t* indica que o conjunto com menor média possui uma menor tendência para desenvolver a doença. Assim, nesse âmbito,
- Pessoas com IMC alto possuem maior tendência para desenvolver *diabetes mellitus*;
- Pessoas com IMC baixo possuem menor tendência para desenvolver *diabetes mellitus*.

```python
# Separando os conjuntos
zero_glu = data_new[data_new["Outcome"] == 0]["Glucose"]
one_glu = data_new[data_new["Outcome"] == 1]["Glucose"]

q1 = one_glu.quantile(0.25) # 1° quartil do conjunto com Outcome = 1
q3 = zero_glu.quantile(0.75) # 3° quartil do conjunto com Outcome = 0

print(f'Valor min: {q1}\nValor máx: {q3}')
```

Isso significa que a faixa em questão está compreendida entre um valor aproximado de 119 *mg/dL* e 125 *mg/dL*. Dentro dessa faixa, há uma grande chance de estar o ponto crítico de interesse. Para aproximá-lo, utiliza-se da curva ROC, a qual compara a sensibilidade com especificidade dos conjuntos em questão. Seu formato permite encontrar um ponto que melhor separa as duas classes. A biblioteca *roc_curve* facilita os cálculos e encontra o valor ótimo de forma simples:

```python
from sklearn.metrics import roc_curve, auc # Bibliotecas para os cálculos da curva ROC

# Separando os conjuntos para o cálculo dos pontos
y_true = data_new['Outcome']
y_scores = data_new['Glucose']

# Calculando os pontos da curva ROC
fpr, tpr, thresholds = roc_curve(y_true, y_scores)

# Calculando o Índice para cada threshold e identificando o melhor valor
youden_index = tpr - fpr
optimal_idx = np.argmax(youden_index)
optimal_threshold = thresholds[optimal_idx]

print("Valor crítico de glicose:", optimal_threshold)
```

Portando, o valor crítico aproximado pela curva ROC é 124 *mg/dL*.

```python
test_size = 0.2 # Proporção de dados de teste em relação ao tamanho do dataset
random_state = 42 # Como o modelo aleatorizará o dataset
solver = 'lbfgs' # Solver que será utilizado pelo modelo
max_iter = 100 # Número de iterações
```

```python
# Importando o modelo de regressão logística do sklearn
from sklearn.linear_model import LogisticRegression
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size=test_size, random_state=random_state)

# Instanciando e treinando o modelo
model = LogisticRegression(random_state=random_state,
                           solver=solver,
                           max_iter=max_iter)
model.fit(x_train, y_train)

# Testando e avaliando o modelo
y_pred = model.predict(x_test)

# Cálculo das métricas
acuracia = accuracy_score(y_test, y_pred)
print(f"Acurácia: {acuracia:}\n")
print(classification_report(y_test, y_pred))
```

No modelo de regressão logística, as duas principais variáveis que trarão diferenças significativas no desempenho do modelo são o *solver* e o *max_iter*. Cabe ao usuário testar as diferentes combinações para chegar a uma acurácia ótima.

Com o modelo treinado, a variável *sample_imput* simula situações de teste para que o modelo proponha uma probabilidade.

```python
sample_input = x_test[0:1]

probabilidades = model.predict_proba(sample_input)

print("Probabilidade de não ter diabetes:", round(probabilidades[0][0],4))
print("Probabilidade de ter diabetes:", round(probabilidades[0][1],4))
```
